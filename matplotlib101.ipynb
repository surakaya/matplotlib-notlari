```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xnoktalari = np.array([0,6])
ynoktalari = np.array([0,100])

plt.plot(xnoktalari,ynoktalari, '-')
plt.show()
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints,marker='*',ms = 20,mec='r',linestyle = 'dotted') #ms=marker size, mec =edge
plt.show()
```

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])
plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
y3 = np.array([1,2,3,4])
y4 = np.array([7,6,5,4])

plt.plot(y3, c = 'black')
plt.plot(y4, c = 'green')
plt.xlabel('x ekseni')
plt.ylabel('y ekseni')
plt.title('grafik başlığı', loc = 'left')
plt.show()
```

```python
y5 = np.array([1,2,3,4])
y6 = np.array([7,6,5,4])

plt.plot(y5, c = 'black')
plt.plot(y6, c = 'green')
plt.xlabel('x ekseni')
plt.ylabel('y ekseni')
plt.title('grafik başlığı', loc = 'left')

plt.grid(color = 'red', linestyle = '--', linewidth = 0.5)
plt.show()
```

**Subplot:**

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) 
#(satır,sütun,1.grafiğin çizimi)
plt.plot(x,y)
plt.title('Grafik 1')

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title('Grafik 2')

plt.suptitle('GRAFIKLER')
plt.show()
```

**ScatterPlot**

Sıcaklık dağılım grafiğidir.


```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y,color = 'red')


x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y,color='black',alpha=0.9)

plt.show()
```

**Barplot**

Çubuk grafiğidir.

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y,width = 0.5,color = 'hotpink')
plt.show()
```

**Histogram**

Frekans dağılım grafiğidir.

```python
x = np.random.normal(170,10,20)
x
```

```python
plt.hist(x,color='purple')
plt.xlabel('Boy (cm)')
plt.ylabel('Yaş')
plt.show()
```

**Pie Charts**

```python
Pasta grafiği çizmemize olanak tanıyan türdür.
```

```python
y = np.array([25,50,75,100])
plt.pie(y)
plt.show()
```

```python
z = np.array([25,50,75,100])
mylabels = ['Honda', 'BMW','Mercedes','Audi' ] #dilimlerin ne anlam ifade ettiğini belirlemek:
plt.pie(z, labels=mylabels, shadow = True)
plt.legend(title = 'Arabalar')
plt.show()
```
